```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Pertama kita import semua Library yang dibutuhkan untuk visualisasi data,lalu kita beri variabel untuk read csv kita,kemudian beri 1 variabel lagi untuk menampilkan pengelompokan datanya

```python
df=pd.read_csv("Weather Data.csv")
data=df.groupby("Weather").sum()
data
```

Bar chart
Adalah diagram batang untuk menampilkan perbandingan antara 2 kolom,disini saya membuat perbandingan,dimana pada suhu tertentu akan terjadi suatu cuaca.

```python
x=data.index 
y=data.Temp_C
plt.bar(x,y)
plt.xlabel("Cuaca")
plt.ylabel("Suhu")
plt.show()
```

Hist plot
Plot untuk menampilkan frekuensi atau seberapa sering kemunculan data tersebut,seperti yang saya tampilkan,frekuensi/kemunculan kondisi cuaca berbeda beda.

```python
plt.hist(df["Weather"])
plt.title("Cuaca")
plt.xlabel("Kondisi cuaca")
plt.ylabel("Frekuensi")
plt.show
```

Pie plot
Grafik lingkaran yang menampilkan persentase banyaknya suatu data dari keseluruhan/100% data dan jangan lupa value counts dulu untuk dihitung

```python
data=df.Weather.value_counts()
plt.pie(data,labels=data.index,autopct="%1.1f%%")
plt.title("Cuaca")
plt.show()
```

Scatter plot
Grafik yang menampilkan hubungan,disini saya menggunakan hubungan dari temperatur dengan kecepatan angin

```python
plt.scatter(df["Wind Speed_km/h"],df["Temp_C"])
plt.title("Suhu")
plt.xlabel("Wind speed/Kecepatan angin")
plt.ylabel("Temperatur/suhu")
plt.show()
```

Line Plot
Grafik garis yang menunjukan time series,disini grafik saya menunjukan bahwa di waktu tertentu,maka suhunya segitu

```python
plt.plot(df["Date/Time"],df["Temp_C"])
plt.title("Suhu")
plt.xlabel("Waktu")
plt.ylabel("Temperatur")
plt.show()
```

Box plot
adalah grafik box yang menunjukan median,disini saya menunjukan median suhu/temperatur
disini juga tidak ada data anomali,karena kalau ada,dia akan muncul titik titik

```python
sns.boxplot(x=df["Temp_C"])
plt.show()
```

Heatmap
Disini kita perlu meng import seaborn  
disini saya menggunakan iloc untuk mengambil index yang berupa angka
Heatmap menunjukan seberapa erat hubungan antara data data numerik kita
misal temperatur dengan dew point sangat erat,tetapi rel hum sangat berlawanan dengan visibility
sementara itu cermin diagonal bernilai 1 karena merupakan dirinya sendiri

```python
cor_matrix=df.iloc[0:,1:7].corr()
sns.heatmap(cor_matrix,cmap="coolwarm",annot=True)
plt.show()
```
